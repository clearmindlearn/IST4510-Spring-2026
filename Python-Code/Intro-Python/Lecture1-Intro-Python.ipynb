# Math Operators

```python
a = 10
b = 3
```

```python
print("a+b", a+b)
print("a-b", a-b)
print("a*b", a*b) # multiplicaiton
print("a/b", a/b)  #division
print("a%b", a%b)  #modular --> remainder
print("a//b", a//b) #floor division operator --> it divides and then rounds down to the nearest integer
print("a**b", a**b) #10^3 = 10*10*10 = 1000
```

#Comparison Operator


*   Boolean --> True or False
*   Boolean Expression will be evaluated to be either True or False



```python
a = 5
b = 8
```

```python
print("a == b", a == b) # a is equal to b ?
print("a != b", a != b) # a is NOT equal to b?
print ("a < b", a < b) # a is less than b?
print ("a > b", a > b) # a is greater than b?
print ("a <= b", a <= b) # a is less than or equal to b?
print( "a >= b", a >= b) # a is greater than or equal to b
```

#Boolean Operator
*  and
*  or
* not

```python
a = True
b = False
```

```python
# and operator
print ("a and b", a and b) # False;  For "a and b" to be true, both a and b must be true
```

```python
# or operator
print ("a or b", a or b) # True;   For "a or b" to be true; only one of them can be True
```

```python
print ("not a:", not a)
```
