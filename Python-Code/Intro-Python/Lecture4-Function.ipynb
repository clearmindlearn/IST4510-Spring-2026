## Part 1 — Quick review: built-in functions for lists

We already used lists, so here is a short review of a few built-in functions you can use **without writing loops**.

````python
# Try running this cell
nums   = [3, 1, 4, 1, 5, 9]
prices = [9.99, 3.50, 4.00]
words  = ["cat", "elephant", "dog", "bee"]
````

### 1) `len()`
`len(x)` returns **how many items** are inside `x`.
- List → number of elements
- String → number of characters


````python
len(nums), len(words), len("Python")
````

**Common pitfall:** the last index in a list is `len(list) - 1`.

````python
nums   = [3, 1, 4, 1, 5, 9]

last_index = len(nums) - 1
print("Last index:", last_index)
print("Last value:", nums[last_index])
````

### 2) `sum()`, `min()`, `max()`
- `sum(numbers)` adds up values
- `min(list)` returns the smallest value
- `max(list)` returns the largest value


````python
nums   = [3, 1, 4, 1, 5, 9]
sum(nums), min(nums), max(nums)
````

### 3) `sorted()`
`sorted(list)` returns a **new** sorted list.
- It does **not** change the original list.
- Use `reverse=True` for descending order.


````python
nums   = [3, 1, 4, 1, 5, 9]

print("Original nums:", nums)
print("sorted(nums):", sorted(nums))
print("sorted(nums, reverse=True):", sorted(nums, reverse=True))
print("Still original nums:", nums)
````

## Part 2 — Writing your own functions (`def`)

Built-in functions are provided by Python.
Now we will write our **own** functions so we can reuse code and keep programs organized.

## Important reminder: indentation matters
Just like `if` statements, **functions use indentation** to define their code block.

Python does **not** use `{}` braces for blocks.
Instead, Python uses **indentation** (usually 4 spaces).

```python
def my_function():
    # this is inside the function
    print("Hello")

# this is outside the function
print("Done")
```


### 1) Defining and calling a function

Basic syntax:

```python
def function_name(parameters):
    # function body (indented)
    return value   # optional
```

- The names in the parentheses are **parameters**.
- The values you pass when you call the function are **arguments**.


````python
# Example 1: defining and calling a function

def add_two_numbers(x, y):
    # x and y are PARAMETERS
    total = x + y
    return total

# 10 and 5 are ARGUMENTS
result = add_two_numbers(10, 5)
print("Result:", result)
````

✅ In the example above:

- `x`, `y` are **parameters** (they live inside the function definition)
- `10`, `5` are **arguments** (they are the actual values you pass when calling the function)


### 2) Return values (`return`)

`return` is a Python keyword.

- `return` **sends a value back** to the caller.
- If a function has **no** `return`, it returns `None`.
- A `return` statement also **ends** the function immediately.

**Important idea:** `print()` shows something on the screen, but `return` gives you a value you can store and use later.


````python
# Try running this cell

def square(n):
    return n * n

def print_square(n):
    print(n * n)   # prints, but does not return a value

a = square(4)
b = print_square(4)

print("a =", a)  # 16
print("b =", b)  # None
````

### Example: using `if` inside a function
Functions often contain `if` statements.

Here is a simple **sign** function:

- returns `-1` if `x` is negative
- returns `1` if `x` is positive
- returns `0` if `x` is zero


````python
def sign(x):
    if x < 0:
        return -1
    elif x > 0:
        return 1
    else:
        return 0

print(sign(-3))
print(sign(10))
print(sign(0))
````

### 3) Positional arguments vs keyword arguments

When calling a function, Python matches arguments to parameters in two common ways:

1. **Positional arguments**: matched by **position**
2. **Keyword arguments**: matched by **name**

Keyword arguments are helpful when:
- a function has many parameters, or
- you want your call to be extra readable.


### Rules for the positional and keyword arguments:
Psitional arguments must come first.

Once you use any keyword argument (like b=2), every argument after that must also be a keyword argument (you cannot put a positional argument after a keyword argument).

````python
def order_ice_cream(flavor, size, topping):
    return f"{size} {flavor} ice cream with {topping}"

# Positional arguments (matched by position)
print(order_ice_cream("vanilla", "large", "sprinkles"))

# Keyword arguments (matched by name)
print(order_ice_cream(topping="sprinkles", size="large", flavor="vanilla"))

# Mixing: positional first, then keyword
print(order_ice_cream("vanilla", size="large", topping="sprinkles"))

# Error!!!!!!! Arguments after a keyword argument must be a keyword argument
#print(order_ice_cream("vanilla", size = "large", "sprinkles"))
````

**Rule:** positional arguments must come **before** keyword arguments.

✅ OK: `order_ice_cream("vanilla", size="large", topping="...")`  
❌ Not OK: `order_ice_cream(flavor="vanilla", "large", "...")` --> flavor is a keyword argument. Thus all the arguments after flavor must also be a keyword argument.


### 4) Default parameters

You can give a parameter a **default value**.
If an argument is not provided, the default value is used.

**Rule:** parameters with defaults must come **after** parameters without defaults.


````python
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

print(greet("Kevin"))                 # uses default greeting
print(greet("Kevin", "Good morning")) # overrides the default
print(greet(name="Kevin"))            # keyword call
print(greet(name="Kevin", greeting="Hi"))
````

## Scope: Variables inside vs outside a function

In Python, function parameters are **local** to the function.

A key detail:
- If the input is a **mutable** object (like a list) and you **modify** it inside the function, the change is visible outside.
- If you **reassign** the parameter name inside the function, it only changes the **local** variable inside the function.


````python
# Scope + side effects (mutable vs immutable)

# Example 1: A list is mutable.
# If we MODIFY the list inside the function, the change is visible outside.

def foo(x):
    x[0] = 100   # modifies the original list object

x = [0, 1, 2, 3]
foo(x)
print("After foo(x), x =", x)   # [100, 1, 2, 3]


# Example 2: An integer is immutable.
# If we REASSIGN the parameter inside the function, it does NOT change the outside variable.

def foo2(x):
    x = 1000     # creates a new local x (reassignment)

x = 0
foo2(x)
print("After foo2(x), x =", x)  # still 0
````
